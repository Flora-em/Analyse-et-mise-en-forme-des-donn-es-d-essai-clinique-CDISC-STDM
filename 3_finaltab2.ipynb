# Analyse statistique : Comparaison de l'efficacité BUP/NA vs CLON

Les critères de jugement évalués (rétention et abstinence) sont des **variables binaires** :
- **Rétention** : le patient est présent (succès = 1) ou absent (échec = 0) à VISITNUM > 12
- **Abstinence** : le test urinaire aux opioïdes est négatif (succès = 1) ou positif (échec = 0)

Ces variables peuvent être résumées par des **proportions de succès** dans chaque bras de traitement. L’objectif est donc de comparer des **proportions entre deux groupes indépendants**.

## Choix du test statistique

Pour comparer l’efficacité des deux traitements (**BUP/NA vs CLON**), nous avons choisi d’utiliser **le test du Khi-deux (χ²) d’indépendance**.

Le test du Khi-deux permet de comparer les **distributions de fréquences** (succès / échec) entre deux groupes indépendants. Il est particulièrement adapté à l’analyse de **variables qualitatives binaires**, comme la rétention et l’abstinence.

Ce test permet d’évaluer s’il existe une **association statistiquement significative** entre le type de traitement administré et la probabilité de succès.
<!--  -->
## Condition d'application du test Khi-2
Les conditions d’utilisation du test du Khi-deux sont respectées dans notre étude :
- Variables qualitatives binaires (succès / échec)
- Groupes indépendants (BUP/NA vs CLON)
- Effectifs suffisants dans chaque cellule (effectifs attendus ≥ 5)

En cas d’effectifs insuffisants dans certaines cellules, un **test exact de Fisher** pourrait être envisagé en alternative.

## Taille des échantillons
Dans notre cas, nos échantillons sont suffisamment grands :
- **Rétention** : n₁ = 233 (BUP/NA), n₂ = 110 (CLON)
- **Abstinence** : n₁ = 227 (BUP/NA), n₂ = 96 (CLON)

Ces tailles d’échantillon garantissent une bonne validité asymptotique du test.
## Hypothèses du test t
**Hypothèses du test de Khi-deus :**<br>
Soit :
- p1 ​: probabilité de succès dans le groupe BUP/NA.
- p2 ​: probabilité de succès dans le groupe CLON.

#### Hypothèse nulle (H0)

- H0 : Les proportions sont égales ($p1 = p2$)

L’hypothèse nulle stipule que la probabilité de succès (rétention ou abstinence) est identique dans les deux groupes, c’est-à-dire que le traitement par BUP/NA n’est ni supérieur ni inférieur à la clonidine.

#### Hypothèse alternative 

- H1 : Les proportions sont différentes ($p1 ≠ p2$)

L’hypothèse alternative stipule qu’il existe une différence entre les deux groupes de traitement. Il s’agit d’un **test bilatéral**, sans hypothèse directionnelle a priori.

## Mesure de l'effet et intervalle de confiance

En complément du test Khi-deux, l'effet du traitement sera quantifié par : 
- La différence de proportions : $Δ=p1​−p2$
- Son intervalle de confiances à 95%.<br>
Pour calculer son intervalle, nous appliquons la formule suivant : 

$$
IC = (p_1 - p_2) \pm Z_{1-\alpha/2} \sqrt{\frac{p_1(1-p_1)}{n_1} + \frac{p_2(1-p_2)}{n_2}}
$$



Ces mesures permettent d’évaluer l’ampleur clinique de l’effet observé, au-delà de la seule significativité statistique

## Seuil de signification
Nous nous plaçons à un niveau d'erreur de : **α = 0,05** (5%).<br>

Si la valeur de p issue du test du Khi-deux est inférieure à 0,05, l’hypothèse nulle sera rejetée et la différence observée sera considérée comme **statistiquement significative**.

# TEST KHI-2

### Test pour la rétention

```python
# Table de contigence : Rétention 
table_ret = np.array([
    [227, 6],    # BUP/NAL: retenus, perdus
    [96, 14]     # CLON: retenus, perdus
])
```

```python
ret = sp.stats.chi2_contingency(table_ret)
ret
```

source : https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.chi2_contingency.html 

- statistic: Plus cette valeur est élevée, plus il y a de différence entre les fréquences observées et attendues, donc plus l’association entre les variables est probable.
- pvalue : C’est la probabilité de trouver une différence aussi extrême (ou plus) si l’hypothèse nulle est vraie. 
- dof : degré de liberté du test.
- expected_freq : ce sont les fréquences attendues sous l’hypothèse d’indépendance.

Nous calculons par la suite manuellement l'intervalle de confiance de la rétention qui encadre l’effet estimé, indique la précision de l’estimation et aide à interpréter sa signification pratique et clinique.

```python
n_BUP = 233
n_CLON = 110
p_BUP = 227/233
p_CLON = 96/110
diff_ret = p_BUP - p_CLON
diff_ret
```

```python
se_ret = np.sqrt((p_BUP*(1-p_BUP)/n_BUP) + (p_CLON*(1-p_CLON)/n_CLON))
se_ret
```

source : https://sites.google.com/view/aide-python/statistiques/moyennes-et-intervalles-de-confiance-en-python

```python
ic_ret = (diff_ret - 1.96*se_ret, diff_ret + 1.96*se_ret)
ic_ret 
```

#### Interprétation :
- χ² = 12.24 → grande différence observée vs attendue (différence de groupe)
- p = 0.000468 → ici p < 0.05 → on rejette l’hypothèse nulle donc il existe une **différence significative** entre nos variables. (la probabilité que cette différence soit due au hasard est très faible)
- - IC 95% : [0.036, 0.167]
  - en % : [3.6%, 16.7%]
  - on peut dire avec 95 % de confiance que BUP/NA améliore la rétention de 3,6 % à 16,7 % par rapport à CLON.

### Test pour l'abstinence

```python
# Table de contigence : Abstinence

table_abs = np.array([
    [191, 36],  # BUP/NA : abstinents, non-abstinents
    [54, 42]    # CLON   : abstinents, non-abstinents
])
```

```python
abs = sp.stats.chi2_contingency(table_abs)
abs
```

```python
#Calcule de IC95% pour l'abstinence

n2_BUP = 227      # Patients BUP/NAL retenus
n2_CLON = 96      # Patients CLON retenus
p2_BUP = 191/227  # Abstinents BUP/NAL
p2_CLON = 54/96   # Abstinents CLON
diff_abs = p2_BUP - p2_CLON
```

```python
se_abs = np.sqrt((p2_BUP*(1-p2_BUP)/n2_BUP) + (p2_CLON*(1-p2_CLON)/n2_CLON))
se_abs 
```

```python
ic_abs = (diff_abs - 1.96*se_abs , diff_abs + 1.96*se_abs )
ic_abs
```

#### Interprétation : 
- χ² = 27.1 → différence observée vs attendue
- p = 1.88e-07 → ici p < 0.05 → on rejette l’hypothèse nulle donc il existe une **différence très significative** entre nos variables.
- - IC 95% : [0.169, 0.389]
  - en % : [16,9%, 38.9%]
  - on peut dire avec 95 % de confiance que BUP/NA améliore l’abstinence de 16,9% à 38.9% par rapport à CLON.

CONCLUSION : Dans les deux cas (rétention et abstinence), nous rejetons H0, ce qui signifie que le traitement au BUP/NA est significativement plus efficace que le traitement CLON.
